## Data Wrangling

Data wrangling is defined as the process of cleaning, structuring, and enriching raw data into the required format to make further analysis and judgements (Dhand, G. et al.20 23). The following steps will include the creation of new columns that will aid the study in the formation of visualizaions. The columns are created through certain calculations based on the existing columns.

```python
#Combines the rank positions across spotify and Apple music for both playlists and charts.
#Helps in providing a unified metric to compare an artists overall popularity across platforms.
def calculate_average_rank(df3): #Calculates average by adding and dividing by the total number.
    df3['Average Rank across platforms'] = (df3['Rank in Spotify playlists'] + 
                          df3['Rank in Spotify charts'] + 
                          df3['Rank in Apple Music playlists'] + 
                          df3['Rank in Apple Music charts']) / 4
    return df3
df3 = calculate_average_rank(df3) #Calls the action done above
df3.head()
```

```python
#Highlights the discrepencies in popularity between two platforms, showing where the artists are more favoured.
def calculate_platform_rank_difference(df3):
    df3['Platform Rank Difference'] = df3['Rank in Apple Music charts'] - df3['Rank in Spotify charts']  # Calculate the rank difference between Spotify and Apple Music charts
    
    return df3
df3 = calculate_platform_rank_difference(df3) #calls the above action
df3.head(20)
```

```python
print(df3.dtypes)
```

```python
def calculate_streaming_bias(df3):
    df3['Streaming Bias'] = df3['Rank in Spotify charts'] - df3['Rank in Apple Music charts']
    return df3
    
df3= calculate_streaming_bias(df3)
df3.head()
```

```python

#This new column measures how consistently an artist(s) ranks across different sections such as the playlists and charts with each platform.
#High consistency = similar rank across both platforms
#low consistency = significant fluctuations
def calculate_platform_rank_consistency(df3):
    df3['Spotify Rank Consistency'] = abs(df3['Rank in Spotify playlists'] - df3['Rank in Spotify charts']) #Calculates absolute differences
    df3['Apple Music Rank Consistency'] = abs(df3['Rank in Apple Music playlists'] - df3['Rank in Apple Music charts'])
    
    df3['Platform Rank Consistency'] = (df3['Spotify Rank Consistency'] + df3['Apple Music Rank Consistency']) / 2 #Calculates the average consistency scores
    df3.drop(['Spotify Rank Consistency', 'Apple Music Rank Consistency'], axis=1, inplace=True) #axis = 1 focuses on the columns whereas 0 denotes rows (McKinney,2022).
    
    return df3
df3= calculate_platform_rank_consistency(df3) #calls the action done above
df3.head()
```

```python

#It becomes difficult for the reader to determine what value can be considerd high and low since there is no measurement scale provided.
print(df3['Platform Rank Consistency'].describe()) #Helps determine max and minimum values and categorises values into quartiles.
```

```python
#For measurement reference values have been assigned weights such as 'very low', 'low', 'Medium', and 'High'.
#The values have been taken from the above shows 'describe' function.
df3['Consistency Scale'] = pd.cut(
    df3['Platform Rank Consistency'],
    bins=[-float('inf'), 3.5, 10.5, 36.5, float('inf')],
    labels=['Very Low', 'Low', 'Medium', 'High']
)

df3
```

```python

#The formula below creates a new column that identifies emerging artists based on the data derived from "year of release" and "Average rank across platforms column".
#Emerging artist in this research will be defined as an artists who debuted in 2024 or if there rank improves significantly between 2023 and 2024.
#Weights have been allocated to both emerging and established artists. 1= "Emerging artist" & 0= "Established artist".
#The main reason of assigning weights rather than simply writing "Emerging" or "Estabished" is so that the columnn type remains float. This will aid and make the visualisations reader friendly.

df3['Emerging Artist'] = df3.apply(
    lambda row: 1 if (row['Year of Release'] == 2024 or 
                      (row['Average Rank across platforms'] < df3[(df3['Year of Release'] == 2023) & 
                                                                 (df3['Artist(s)'] == row['Artist(s)'])]['Average Rank across platforms'].mean())) 
              else 0, 
    axis=1
)
df3.head() 
```

# Data visualization 


The next steps after Data loading and cleaning included Data wrangling where some new columns were created based on existing data. This will be followed by Data visualization for easy interpretation and analysis. Islam and Jin (2019) define Data visualization as "presenation of quantiative information in a graphical form." It turns large data into smaller understandable visuals that are easy for the human brain to comprehend.

A **bar chart** is defined as a chart that uses rectangular bars to represent values of different data categories. The respective height of the bars will be directly proportional to the values it presents. The main purpose of using a bar chart is to draw a comparison across various variables. Bar charts are highly useful in data visualizations because they are a clear way of comparing quantities and highlighting differences and trends across categories (Barker, T. 2013).

The following bar chart creates a grouping of Artist(s) and platform rank difference. It calculates the mean of absolute values and identifies the top 10 artists with the largest average platform rank difference. This chart aids the research in identifying rank discrepencies between Spotify and Apple Music. The height of each bar relates to the magnitude of differences making the comparison easy to visualise. 

This chart aligns directly with the objective of  drawing platform comparison and artist trends. Moreover, it provides an artist centric approach focusing on platfrom dynamics and performance across those platforms.

```python
top_diff_artists = df3.groupby('Artist(s)')['Platform Rank Difference'].mean().abs().nlargest(10)
top_diff_artists.plot(kind='bar', figsize=(10, 6),color= '#669900' ,title='Platform Rank Difference for Top Artists') #figsize defines the figures aspect ratio when saved (McKinney,2022).
plt.ylabel('Rank Difference',color= 'Red')
plt.xlabel('Artist(s)',color='blue')
plt.savefig('Platform_rank_difference_TopArtists.png', dpi=300, bbox_inches='tight') #dpi defines figure ratio in dots per inch. It is configurable depending on the user (McKinney,2022).
plt.show()
```

A **stacked bar chart** is similar to a bar chart except that it divides each bar into sub-categories. It is used to depict proportions of subcategories within the categories. It is useful in data visualization as it adds depth to the chart by showing various proportions of data while comparing the total at the same time (Barker, T. 2013). The stacked bar chart below depicts a comparison of platform rank consistency for emerging and established artists by year. It aids the research in analyzing the trends for emerging artists, whether they gained prominence in recent years or if established artists still dominate.

```python

# Group data and calculate count of 'Platform Rank Consistency' for each combination of 'Year of Release' and 'Emerging Artist'
consistency_count = df3.groupby(['Year of Release', 'Emerging Artist'])['Platform Rank Consistency'].count().unstack(fill_value=0)
ax = consistency_count.plot(
    kind='bar', 
    stacked=True, 
    figsize=(10, 6), 
    title='Rank Consistency: Emerging vs. Established Artists'
)

plt.ylabel('Number of Songs')  # Label for y-axis
plt.xlabel('Year of Release')  # Label for x-axis
plt.xticks(rotation=0)  # Rotate x-axis ticks for readability
plt.legend(title='Artist Type', labels=['Established', 'Emerging'])
plt.savefig('Rank_consistency_emerging_established.png', dpi=300, bbox_inches='tight')
plt.show()
```

The **stacked bar chart** below shows the distribution of songs for emerging and established artists. Each bar is divided into consistency scale with a color key indicating the meaning of each color. The x-axis differentiates between artist types and the y-axis shows the number of songs.

The visualisation below aids the reader and the researcher in understanding trends in consistency, whether emerging artists tend to have higher or lower consistency than established artists. Moreover, it also provides a comparative analysis and allows the researcher to analyse if emerging artists are "closing the gap" in consistency or whether the established artists withold their scales.

This visualisation help the research to achieve the objective of identifying emerging artists and understanding dynamics related to them. This visualisation differs from the previously shown stacked bar chart. The previous visualisation showed the rank consistency by year of release, wherease this one shows rank consistency by emerging or established artists. This visualisation helps aid the previous visulaisation in understanding rank consistency and artist(s) dynamics.

```python
consistency_scale_artist = df3.groupby(['Emerging Artist', 'Consistency Scale']).size().unstack(fill_value=0)
consistency_scale_artist.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis', title='Consistency Scale: Emerging vs. Established Artists')
plt.xlabel('Artist Type (0: Established, 1: Emerging)')
plt.ylabel('Number of Songs')
plt.legend(title='Consistency Scale')
plt.xticks(rotation=0)
plt.show()
```

This **bar chart** indicates a simple yet insightful comparison of platform preferences.

This shows how popularity of an artist may also contribute towards the success of a platform thus opening ways for future artists as well to have themselves featured on Spotify.

```python
artist_preference = df3.groupby(['Artist(s)', 'Preferred Platform']).size().unstack(fill_value=0) #groupby function allows to slice ,dice, and summarize datasets (McKinney,2022).
top_artists = artist_preference.sum(axis=1).nlargest(10).index
artist_preference.loc[top_artists].plot(kind='bar', figsize=(12, 8), color=['#1cdcec','#1ED760'], title='Top 10 Artists by Platform Preference')
plt.xlabel('Artist(s) (Top 10)')
plt.ylabel('Number of Songs')
plt.legend(title='Preferred Platform')
plt.xticks(rotation=45)
plt.savefig('Top_10_Artists', dpi=300, bbox_inches='tight')
plt.show()
```

A **heatmap** visualizes a color-based approach to data representation with two distinct variables being associated to the axis’s and a third variable that is color coded (Cremonini, M. 2024). Their use in large data sets is helpful as they identify high and low values along with highlighting correlations that may require further analysis. The information displayed by a heatmap is highly qualitative with the color gradient having quantitative values (Cremonini, M. 2024). Heatmaps are used in this study to depict rank differences or consistencies. They have also been used to compare rank consistencies of emerging and established artists.

The heatmap titled "Average Rank Difference by Preferred Platform and Year" provides an insight on how the average platform rank difference varies across Spotify and Apple Music for 2023 and 2024. The rows display Spotify and Apple Music as the prefered platform wherease teh columns differentiate between Year of release. The color intensity represents the magnitude of the rank difference. Warmer color indicates a higher difference while cooler colors indicate lower difference. 

The visualisation aids the researche in understanding platform- specific insights along with yearly trends in platform preference. It also aids in achieving the objective of understanding platform popularity by displaying where rank perceptions differ the most.


```python
pivot_table = df3.pivot_table(
    index='Preferred Platform',
    columns='Year of Release',
    values='Platform Rank Difference',
    aggfunc='mean'
)
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', cbar=True)
plt.title('Average Rank Difference by Preferred Platform and Year')
plt.xlabel('Year of Release')
plt.ylabel('Preferred Platform')
plt.savefig('Average_Rank_difference_platform&Year.png', dpi=300, bbox_inches='tight')
plt.show()
```

They are primarily used to analyze the distribution, skewness, and range of variables in the given data (Cremonini, M. 2024). Histograms have been used in this study for visualization purposes of identifying trends in consistency scale and platform rank difference.

The code below creates a histogram to visualise the distribution of the consistency scale. 
By analysing the shape of the histogram it may be determined how many artists fall into specific ranges of consistency scale values.

A **violin plot** is a combination of a *box plot* and *kernel density plot* showcasing data distribution summary statistics. The shape of the violin is adapted through the density of data points for each value of the variable. An abundance of data points will make the shape larger while the scarcity of them will make it smaller (Cremonini, M. 2024). Violin plot has been used in this study to analyze platform rank consistency across streaming bias categories. 

The following visual depicts a violin plot and examines the relationship between streaming bias and platform rank consistency. 
The shape of the violin indicates variability patterns. It highlights if consistency varies significantly between different bias categories. 
This visual, aids in the analysis of third objective by providing insights into how bias affects rank consistency showing whether Spotify or Apple Music has more stable rankings.
In the limited capacity of this research, the purpose of identifying biases was to understand whether rank consistencies occured due to varying biases.

```python
# Creates bins for better visualisation and understanding. The original data contains negative and positive values. If the values are positive then "Spotify biases will appear"
#If negative then "Apple biased" otherwise "unbiased". There are no zero values therefore the graph shows a flat line above the unbiased column.
df3['Streaming Bias Category'] = df3['Streaming Bias'].apply(lambda x: 
                                    'Spotify-Biased' if x > 0 
                                    else 'Apple-Biased' if x < 0 
                                    else 'Unbiased')

# Creates a violin plot
sns.violinplot(x='Streaming Bias Category', y='Platform Rank Consistency', data=df3, palette='coolwarm')
plt.title('Distribution of Platform Rank Consistency by Streaming Bias')
plt.xlabel('Streaming Bias')
plt.ylabel('Platform Rank Consistency')
plt.savefig('PlatformRank_StreamingBias.png', dpi=300, bbox_inches='tight')
plt.show()
```

A **scatter plot** represents the relationship between two continuous variables by using points on a two-dimensional graph (cartesian grid). Scatter plots are used to spot correlations, patterns, or clusters between two variables (Barker, T. 2013). They are primarily used to understanding how two variables react against each other. The patterns formed by the points on the scatter plot depict certain relationships. A positive correlation occurs when a uniform increase is noticed in the variables and a negative correlation when an increase in one variable leads to decrease of the other variable (Barker, T. 2013). Scatter plot is used in this study to analyze and visualize patterns linking to streaming bias and consistency scale. 

The scatter plot provides a visual representation of the relationship between platform rank difference and consistency with the consistecy scale providing further clarity through color coding. The axis indicates difference in artist ranks between Spotify abd Apple Music. Negative values depict better performance on Spotify wherease positive values show better performance on Apple Music. As judged by the scatter plot Spotify shows better performance whereas Apple Music shows fewer values and is more prone towards discrepencies. (Cluster of points at a low rank differences and high consistency indicate that samller differences correlate with greater stability)

This scatter plot helps the research by identifying correlations between variables. This visualisation aids in performing a trend analysis(first objective). Although not significantly viewable this graph may also identify emerging artists by identifying specific patterns of rank differences or lower consistency. If an artists has lower rank differences on both platforms and higher consistencies across platforms and years, this would mean that they are an emerging artist.

```python

sns.scatterplot(x='Platform Rank Difference', y='Platform Rank Consistency', hue='Consistency Scale', data=df3, palette='muted')
plt.title('Rank Difference vs. Consistency')
plt.xlabel('Platform Rank Difference')
plt.ylabel('Platform Rank Consistency')
plt.legend(title='Consistency Scale')
plt.savefig('Rank_difference_vs_Consistency.png', dpi=300, bbox_inches='tight')
plt.show()
```

The following **heatmap** titled *Average Rank Consistency Across Platforms by Artist Type* shows the avaerage platform rank consistency for emerging and established artists. A **line plot** connects data points with a line to visualize changes in data or variables over time (Pajankar, A. 2022). Line plots are usually used for time-series data and highlighting trends. Line plots are used in this research to study rank consistencies over time for established and emerging artists. 

The following chart shows a line plot that analyzes the trend of average platform rank consistency over the period of two years (2023 and 2024). The code depicts a grouping of data by year of release and emerging artists and calculates mean platform rank consistency for each specific group. The visualization depicts temporal trends, comparison between artist types and yearly performance. 

The purpose of this visualization was to understand trends in rank consistency over time. 

```python
trend_data = df3.groupby(['Year of Release', 'Emerging Artist'])['Platform Rank Consistency'].mean().reset_index()
sns.lineplot(
    x='Year of Release', 
    y='Platform Rank Consistency', 
    hue='Emerging Artist', 
    data=trend_data, 
    palette='Set2'
)

plt.title('Rank Consistency Trend Over Time')
plt.xlabel('Year of Release')
plt.ylabel('Average Platform Rank Consistency')
plt.xticks(
    ticks=trend_data['Year of Release'].unique(), 
    labels=trend_data['Year of Release'].unique().astype(int), 
    rotation=45  # Optional: Rotate for better readability
)

plt.legend(title='Artist Type', labels=['Established Artists', 'Emerging Artists'])
plt.savefig('Rank_consistency_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Limitations

<br>1. **Algorithmic transparency**: This research assumes that discrepencies in artist rankings and their consistency is influenced by platform algorithms and user preferences. Lack of transparency from the respective platforms and their measuring mechanisms leads to a limitation for this research by not reaching a definitive conclusion on the causality of certain trends. <br>
<br>2. **External factors**: There are certain external factors that may skew the analysis. For example, promotional campaigns, potentiol collaborations, and third party platforms like TikTok shape user preferences and may influence artist popularity. These factors are not directly accounted for and thus may lead to a skewed analysis. <br>
<br>3. **Limited time frame**: The research focuses on data only from 2023 and 2024. This provides a limited time frame for analysis of past and future trends in order to understand shifting user preferences and platform strategies. <br>
<br>4. **Geographical Bias**: The dataset does not focus on regional statistics such as market preferences. Lack of geographical information leads to a limited analysis of rankings and user behaviours. <br>
   

## Managerial implications

<br>1. **Platform Strategy optimization**: Streaming platforms can use this research to gain insights and optimize their algorithms according to market trends.
